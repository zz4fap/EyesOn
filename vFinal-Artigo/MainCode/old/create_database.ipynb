```python
import csv

def convert_to_float(data):
    new_data = []
    for item in data:
        new_item = []
        for subitem in item:
            if 'array' in subitem:
                # Avalia o array do NumPy como uma express√£o
                new_subitem = eval(subitem)
                # Converte para lista de floats
                new_subitem = [float(value) for value in new_subitem]
            else:
                # Converte a string para float diretamente
                new_subitem = float(subitem)
            new_item.append(new_subitem)
        new_data.append(new_item)
    return new_data

#print(data)
#for i in data:
#    print(i)

#pitch, yaw, px, py


#print(all_data)
data3 = []
for i in range(5):
    with open(f'calib_10xPonto/calib10/calibration5s_file10-pt{i+1}.csv', newline='') as f:
        reader = csv.reader(f)
        data = list(reader)
        data2 = [item for item in data if item]
        data2 = convert_to_float(data2)
        data3.append(data2)
        #print(data2[0])

data4 = []


for da in data3:
    for d in da:
        data4.append(d)

    

'''tt = 0
for x in range(16):
    for d in data3:
        print(d)
        #print(d[tt])
        data4.append(d[tt])
    tt+=1'''


with open('calib_10xPonto/calib10/calib_all_file10.csv', 'w', newline='') as f:
    # using csv.writer method from CSV package
    write = csv.writer(f)
    write.writerows(data4)
```
